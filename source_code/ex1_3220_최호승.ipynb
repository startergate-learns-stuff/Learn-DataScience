```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

rc('font', family='AppleGothic')
```

```python
plt.plot(ac['시간'], ac['선형가속도'])
plt.show()
```

```python
plt.plot(ac['시간'], ac['선형가속도'])
plt.title('선형가속도계')
plt.xlabel('시간')
plt.ylabel('선형가속도')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ac['시간'], ac['선형가속도'], color='green', linestyle='dashed', marker='o')
plt.title('선형가속도계')
plt.grid()
plt.xlabel('시간')
plt.ylabel('선형가속도')
plt.show()
```
