```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import rc

plt.rcParams['axes.unicode_minus'] = False

rc('font', family='AppleGothic')
```

```python
t = np.arange(0, 10, 0.1)
y = np.sin(t)
```

```python
plt.plot(t, y)
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
plt.plot(t, y, color = 'r', linestyle=':', linewidth=5, marker='*')
plt.grid()
plt.plot(t, np.cos(t), color='g', linestyle='--', linewidth= 5, marker='o', label = 'cost(t)')
plt.figure(figsize=[6, 7])
plt.title('sin ')
plt.xlabel('x ')
plt.ylabel('y ')
plt.ylim(-2.2)
plt.text(6, 1.5, ' ')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, np.sin(t))

plt.subplot(2, 1, 2)
plt.plot(t, np.cos(t))
plt.show()
```

```python
plt.scatter(t, y, s=5)
plt.show()
```

```python
df = pd.read_csv('../data/suggestion.csv', encoding='utf-8')

del df['Unnamed: 0']
df.head()
```

```python
df.set_index('id', inplace=True)
df.head()
```

```python
plt.barh(df[' '], df[' '])
```

```python
df2 = df.groupby(by=' ')[[' ']].sum()
df2 = df2.sort_values(by=' ', ascending=False)
df2
```

```python
df2[' '].sort_values().plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title('  ')

df2[' '].sort_values().plot(kind='barh', color = 'r', linewidth= 3,)
plt.text(6, 1.5, ' ')
plt.grid()
plt.xlabel(' ')
plt.xlim(0, 50)
plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
#
plt.plot(t, y, color='r', linestyle= ':', linewidth = 5, marker = '+', label = 'sin(t)')
plt.grid() #  
# 
plt.plot(t, np.cos(t), color='g', linestyle='--', linewidth= 5, marker='o', label = 'cost(t)')

plt.title('')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 2)
plt.text(6, 1.5, ' ')
plt.legend() #
plt.show()
```

```python
plt.figure(figsize=(7, 7))

df2[' '].sort_values().plot(kind='barh', color='r')
plt.grid()
plt.title('   ')
plt.xlabel(' ')
plt.xlim(0, 50)

for i in range(3):
    plt.text(df2[' '][i], 5-i, df2[' '][i], fontsize=15)

plt.savefig('  .png')
```
