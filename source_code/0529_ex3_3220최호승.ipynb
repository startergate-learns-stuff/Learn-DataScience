```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import rc

plt.rcParams['axes.unicode_minus'] = False

rc('font', family='AppleGothic')
```

```python
t = np.arange(0, 10, 0.1)
y = np.sin(t)
```

```python
plt.plot(t, y)
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
plt.plot(t, y, color = 'r', linestyle=':', linewidth=5, marker='*')
plt.grid()
plt.plot(t, np.cos(t), color='g', linestyle='--', linewidth= 5, marker='o', label = 'cost(t)')
plt.figure(figsize=[6, 7])
plt.title('sin 함수')
plt.xlabel('x축 좌표')
plt.ylabel('y축 좌표')
plt.ylim(-2.2)
plt.text(6, 1.5, '빅데이터 분석')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, np.sin(t))

plt.subplot(2, 1, 2)
plt.plot(t, np.cos(t))
plt.show()
```

```python
plt.scatter(t, y, s=5)
plt.show()
```

```python
df = pd.read_csv('../data/suggestion.csv', encoding='utf-8')

del df['Unnamed: 0']
df.head()
```

```python
df.set_index('id', inplace=True)
df.head()
```

```python
plt.barh(df['제안 분야'], df['투표 수'])
```

```python
df2 = df.groupby(by='제안 분야')[['투표 수']].sum()
df2 = df2.sort_values(by='투표 수', ascending=False)
df2
```

```python
df2['투표 수'].sort_values().plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title('제안 분야별 특표수')

df2['투표 수'].sort_values().plot(kind='barh', color = 'r', linewidth= 3,)
plt.text(6, 1.5, '빅데이터 분석')
plt.grid()
plt.xlabel('투표 수')
plt.xlim(0, 50)
plt.ylabel('제안 분야')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
#사인함수
plt.plot(t, y, color='r', linestyle= ':', linewidth = 5, marker = '+', label = 'sin(t)')
plt.grid() # 각자 무늬
#코사인 함수
plt.plot(t, np.cos(t), color='g', linestyle='--', linewidth= 5, marker='o', label = 'cost(t)')

plt.title('사인함수')
plt.xlabel('x좌표')
plt.ylabel('y좌표')
plt.ylim(-2, 2)
plt.text(6, 1.5, '빅데이터 분석')
plt.legend() #범례표현
plt.show()
```

```python
plt.figure(figsize=(7, 7))

df2['투표 수'].sort_values().plot(kind='barh', color='r')
plt.grid()
plt.title('제안 분야별 투표 수')
plt.xlabel('투표 수')
plt.xlim(0, 50)

for i in range(3):
    plt.text(df2['투표 수'][i], 5-i, df2['투표 수'][i], fontsize=15)

plt.savefig('제안 분야별 투표율.png')
```
