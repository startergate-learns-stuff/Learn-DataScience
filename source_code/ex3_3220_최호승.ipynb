```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import rc
plt.rcParams['axes.unicode_minus'] = False

rc('font', family='AppleGothic')

plt.figure(figsize=(8, 6))
plt.scatter(iris_Df['sepal length (cm)'], iris_Df['sepal width (cm)'], s=50)

plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
markers = ['^', 's', 'o']
for i, markers in enumerate(markers):
    x_axis_data = iris_Df[iris_Df['target'] == i]['sepal length (cm)']
    y_axis_data = iris_Df[iris_Df['target'] == i]['sepal width (cm)']
    plt.scatter(x_axis_data, y_axis_data, marker=markers, label=iris.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
iris_Df2 = iris_Df.copy()

del iris_Df2['target']
iris_Df2.head(3)
```

```python
iris_Df3 = iris_Df.copy()

iris_Df4 = iris_Df3.drop('target', axis=1)
iris_Df4.head()
```

```python
from sklearn.preprocessing import StandardScaler

iris_scaled = StandardScaler().fit_transform(iris_Df2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(iris_scaled)
print(pca.components_)

iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
iris_pca_Df = pd.DataFrame(iris_pca, columns=['pca1', 'pca2'])
iris_pca_Df.head()
```

```python
iris_pca_Df['target'] = iris.target
iris_pca_Df.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8, 6))

markers = ['^', 's', 'o']
for i, markers in enumerate(markers):
    x_axis_data = iris_pca_Df[iris_Df['target'] == i]['pca1']
    y_axis_data = iris_pca_Df[iris_Df['target'] == i]['pca2']
    plt.scatter(x_axis_data, y_axis_data, marker=markers, label=iris.target_names[i])

plt.legend()
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```
